```python
import os
from os import listdir
from os.path import isfile, join
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
def read_files(path):
    record = {'test': {}, 'train': {}}
    files = [f for f in listdir(path) if isfile(join(path, f))]
    for var in files:
        if 'hs' in var:
            tmp = var[-10:-4]
        else:
            tmp = var[4:10]
        if 'test' in var:
            record['test'][tmp] = pd.read_csv(path+var)['Value'].values
        elif 'train' in var:
            record['train'][tmp] = pd.read_csv(path+var)['Value'].values
    return record
```

```python
def plot_record(record, metric, filename):
    plt.figure(figsize=(10,4))
    for i, opt in enumerate(['train', 'test']):
        plt.subplot(1,2,i+1)
        if 'hs' in filename:
            models = ['hs1000', 'hs2000', 'hs5000']
        else:
            models = ['nmodel', 'bmodel', 'wmodel']
        for model in models:
            if metric == 'accuracy':
                plt.plot(record[opt][model]*100, marker='.', label=f'{model}: {opt}')
                plt.ylabel(metric+' %')
            else:
                plt.plot(record[opt][model], marker='.', label=f'{model}: {opt}')
                plt.ylabel(metric)
        plt.legend()
        plt.xlabel('epoch')
        plt.tight_layout()
        plt.grid()
    plt.savefig('./plots/'+filename)
plt.show()
```

```python
accuracy = read_files('./record/accuracy/')
plot_record(accuracy, 'accuracy', 'diffm_acc.pdf')
```

```python
loss = read_files('./record/loss/')
plot_record(loss, 'loss', 'diffm_loss.pdf')
```

```python
accuracy_hs = read_files('./record/hs/accuracy/')
plot_record(accuracy_hs, 'accuracy', 'diffhs_acc.pdf')
```

```python
loss_hs = read_files('./record/hs/loss/')
plot_record(loss_hs, 'loss', 'diffhs_loss.pdf')
```
